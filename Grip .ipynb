```python
#splitting the dataset into testing and the training datasets 
#import the skicit learn library
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis=1)
Y = df['target']
```

```python
X_train , X_test , Y_train , Y_test =train_test_split(X,Y,test_size=0.3 , random_state = 34)
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
#create a linear model and fitting the model
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train ,Y_train)
```

```python
#predicting the model 
```

```python
 y = model.predict(X_test)
```

```python
#evaluating the performance using mse 
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y,Y_test)
print(mse)
```

```python
#evaluating the performance using r2_score
```

```python
r2 = r2_score(y,Y_test)
print(r2)
```
